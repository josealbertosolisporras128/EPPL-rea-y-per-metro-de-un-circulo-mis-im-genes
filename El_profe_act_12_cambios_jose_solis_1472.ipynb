1. CP --> costo del producto
2. CD --> cantidad de dinero

## formulas
$cambio = CD - CP$

```python
# Calculando el cambipo usando if else
# datos de entrada CD, CP
# mostrar cambio si es falso
# si es verdadero mostar no alcanza

CD = int(input("Ingrese el costo del producto: "))
CP = int(input("Ingrese la cantidad de dinero: "))
if CP < CD:
    print("No alcanza")
else:
    cambio = CD - CP
    print("El cambio es: ", cambio)

print("Fin del programa")
print("programa realizado por jose solis NL = 51")
```
