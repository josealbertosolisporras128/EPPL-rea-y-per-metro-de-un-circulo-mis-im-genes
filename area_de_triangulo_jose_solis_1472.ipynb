```python
# calcular area de un triangulo rectangulo
# variables A, b, h
# formulas A = (b * h) / 2
# capturar el valor de h y b
import math
print("calcular area de un triangulo rectangulo")
b = float(input("ingrese el valor de la altura: "))
h = float(input("ingrese el valor de la base: "))

A = (b * h) / 2
print("el area del triangulo es: ", A)
print("jose solis NL = 51")
```
